```python
import pandas as pd
```

```python
Mahnur_file_location = 'https://github.com/PUBPOL542G3/pythondeliverable-/blob/main/mahnur.xlsx?raw=true'
DF_C02 = pd.read_excel(Mahnur_file_location)
display(DF_C02)
```

```python
Marina_file_location = 'https://github.com/PUBPOL542G3/pythondeliverable-/blob/main/marina.xlsx?raw=true'
DF_ForestArea = pd.read_excel(Marina_file_location)
display(DF_ForestArea)
```

```python
Oliver_file_location = 'https://github.com/PUBPOL542G3/pythondeliverable-/blob/main/OliverRevisedData.xlsx?raw=true'
DF_Employment = pd.read_excel(Oliver_file_location)
display(DF_Employment)
```

```python
DF_Employment.merge(DF_C02,left_on='country',right_on='country',how='left')
```

```python
Oliver_Nur_Merge = DF_Employment.merge(DF_C02,left_on='country',right_on='country',how='outer',indicator='True') 
display(Oliver_Nur_Merge)
```

```python
Oliver_Nur_Merge[Oliver_Nur_Merge['True'] == 'right_only']
```

```python
Oliver_Nur_Merge[Oliver_Nur_Merge['True'] == 'left_only']
```

```python
Triple_Merge = Oliver_Nur_Merge.merge(DF_ForestArea,left_on='country',right_on='country',how='outer',indicator='Check') 
display(Triple_Merge)
```

```python
Left_only_Data = Triple_Merge[Triple_Merge["Check"] == 'left_only']
```

```python
Right_only_Data = Triple_Merge[Triple_Merge["Check"] == 'right_only']
```

```python
Left_only_Data.to_excel('LeftRightData.xlsx')
```

# Oliver Hirn Employment Data Cleaning
In this Jupyter notebook, I will use python code in order to upload, clean and export as an xlsx file data that I pulled from the internet. 



This code imports and reads the "pandas" library package so that I can upload and manipulate files throughout the following code. 

```python
import pandas as pandas
```

Here, I am importing, reading and dispalying the raw employtment data from my Githib repository "PUBPOL542G3/Oliver". I display the data in order to double check that all the data was importing correctly. 

```python
employment_file_location = 'https://github.com/PUBPOL542G3/Oliver/raw/main/RAWDATA_aged_15plus_employment_rate_percent.xlsx'
DF_employment = pandas.read_excel(employment_file_location)
display(DF_employment)
```

After importing the data, the following code narrows down the data that I imported to only show my what I am looking for. In this case, I refined the full data set to only display the country and employment data from the year 2015. I display the data frame after these changes to make sure everything worked properly. 

```python
Refined_Columns = ['country',2015]
DF_employment2015 = DF_employment[Refined_Columns]
display(DF_employment2015)
```

Next, I renamed the column previously labeled as 2015 to "employment". This is so that when I merge my data with my groupmembers 2015 data, we will know what the data represents. I display the renamed column to make certain that everything worked properly. 

```python
#renaming column named 205 to "employment" and then displaying dataframe
DF_employment2015.rename(columns= {2015:'employment'}, inplace= True)
display(DF_employment2015)
```

After renaming the column, I needed to rename a few of the countries in order to make sure that they matched the naming conventions of my groupmates. In this code, I identified which existing country names I wanted to change and then presented the new country name that I wanted to replace the old name with.

```python
#renaming inconsistent country names to match other dataframe naming conventions
DF_employment2015Revised = DF_employment2015.replace({
     'Congo, Dem. Rep.':'Democratic Republic of the Congo',
    'Congo, Rep.': 'Republic of the Congo',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Slovak Republic': 'Slovakia'
})
display(DF_employment2015Revised)
```

This code searches for a specific country name in the data frame that I have created and, if it finds the country, will return the country name and the associated employment data. I ran this code to ensure that the changes I had made to the country names were successful. 

```python
DF_employment2015Revised[DF_employment2015Revised['country'] == 'Slovakia']
```

Here, I am naming the excel file that I will  export this imported and cleaned data set as. I have named the file, "OliverRevisedData.xlsx" 

```python
DF_employment2015Revised.to_excel('OliverRevisedData.xlsx')
```

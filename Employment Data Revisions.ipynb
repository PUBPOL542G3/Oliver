```python
region_file_location = 'https://github.com/PUBPOL542G3/pythondeliverable-/blob/main/Regions.xlsx?raw=true'
DF_Regions = pandas.read_excel(region_file_location,sheet_name='list-of-countries-etc')
display(DF_Regions)
```

```python
Refined_Col_Region = ['name', 'four_regions']
DF_RevRegions = DF_Regions[Refined_Col_Region]
display(DF_RevRegions)
```

```python
DF_RevRegions.rename(columns= {'name':'country'}, inplace= True)
DF_RevRegions.rename(columns= {'four_regions':'region'}, inplace= True)
display(DF_RevRegions)
```

```python
DF_FinRegions = DF_RevRegions.replace({
     'Congo, Dem. Rep.':'Democratic Republic of the Congo',
    'Congo, Rep.': 'Republic of the Congo',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Slovak Republic': 'Slovakia'
})
display(DF_FinRegions)
```
